### Data Loading
Loading the two dataset files "Cycle_and_pedestrian data.csv" and "Weather_data.csv".

```python
df_1 = pd.read_csv("Cycle_and_pedestrian data.csv", header=0)
print(df_1.info())
df_1.head()
```

```python
df_2 = pd.read_csv("Weather_data.csv", header=0)
print(df_2.info())
df_2.head()
```

Convert the values in the Date column (which are in the object data type by default) in both dataframes to datetime objects. This will allow for easier manipulation using datetime properties.

In df_1 the date is given as Sat, 01 01 22 which corresponds to the format "%a, %d %m %y".
In df_2 the date is given as 1-Jan-22 which corresponds to the format "%d-%b-%y".

```python
df_1['Date'] = pd.to_datetime(df_2['Date'], format='%a, %d %m %y')
df_2['Date'] = pd.to_datetime(df_2['Date'], format='%d-%b-%y')
```

```python
print(df_1.info())
print(df_2.info())
```
